```python
import numpy as np
import pandas as pd 
```

```python
#making file
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
movies.shape
```

```python
credits.head(1)
```

```python
#join both on title 
movies = movies.merge(credits, on='title')
```

Determing factors that will allow us to reccommend similar movies: 
- id 
- overview 
- genre
- keywords
- cast 
- crew

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L
```

```python
#remove NA values
movies.dropna(inplace=True)
```

```python
#Convert from Python Dicitonary format to only its key word.  
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
movies['cast'] = movies['cast'].str[:5]
```

```python
movies.head()
```

```python
#Find the director from Dict 
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name']) 
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
movies.rename(columns={'crew': 'director'}, inplace=True)
```

```python
movies.sample(5)
```

```python
def reduce_space(words):
    L = []
    for i in words:
        L.append(i.replace(" ",""))
    return L 
```

```python
movies['cast'] = movies['cast'].apply(reduce_space)
movies['director'] = movies['director'].apply(reduce_space)
movies['genres'] = movies['genres'].apply(reduce_space)
movies['keywords'] = movies['keywords'].apply(reduce_space)
movies['overview'] = movies["overview"].str.split()
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['director']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','director'])
#new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('The Dark Knight')
```
